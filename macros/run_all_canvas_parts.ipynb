```python
import numpy as np
import rplacem.canvas_part as cp
import rplacem.utilities as util
import rplacem.variables_rplace2022 as var
import matplotlib.pyplot as plt
import sys
import os
import glob
import json
import seaborn as sns
import pickle
import timeit
```

```python
delete_png = True # too many files
atlas, atlas_size = util.load_atlas()
```

```python
pixel_changes_all = util.get_all_pixel_changes()
```

```python
starttime = timeit.default_timer()
canvas_comps = []
for i in range(atlas_size):
    print('composition count: ' + str(i + 1))
    atlas_id = atlas[i]['id']
    canvas_comps.append(cp.CanvasPart(atlas_id, 
                        pixel_changes_all=pixel_changes_all,
                        atlas=atlas))
print("The loop time is :", timeit.default_timer() - starttime, "s")
```

```python
file_path = os.path.join(var.DATA_PATH, 'canvas_compositions_all.pickle')
with open(file_path, 'wb') as handle:
    pickle.dump(canvas_comps,
                handle,
                protocol=pickle.HIGHEST_PROTOCOL)
```
