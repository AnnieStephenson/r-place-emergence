```python
import numpy as np
import rplacem.canvas_part as cp
import rplacem.canvas_part_statistics as stat
import rplacem.utilities as util
import rplacem.variables_rplace2022 as var
import rplacem.early_warning_signals as ews_var
import matplotlib.pyplot as plt
import sys
import os
import glob
import json
import seaborn as sns
import pickle
import timeit
```

```python
file_path = os.path.join(var.DATA_PATH, 'canvas_composition_statistics_all_trans.pickle')
with open(file_path, 'rb') as f:
    canvas_comp_stat_list = pickle.load(f)
```

```python
thresh_low = 0.01
thresh_high = 20
time_window = 5
early_warn_signals_list = []

for i in range(len(canvas_comp_stat_list)):
    print('CanvasComp count: ' + str(i+1))
    early_warn_signal = ews_var.EarlyWarnSignals(canvas_comp_stat_list[i],
                                                 time_window,
                                                 state_vars=None,
                                                 half_stab_interval=True,
                                                 thresh_low = 0.01,
                                                 thresh_high = 20)
    early_warn_signals_list.append(early_warn_signal)
```

```python
file_path = os.path.join(var.DATA_PATH, 'early_warn_signals_list.pickle')
with open(file_path, 'wb') as handle:
    pickle.dump(early_warn_signals_list,
                handle,
                protocol=pickle.HIGHEST_PROTOCOL)
```
